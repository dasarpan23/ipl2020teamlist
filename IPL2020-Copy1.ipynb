```python
import requests
from bs4 import BeautifulSoup
```

```python
#IDs of IPL Team
ID = ['1210611','1210601','1210602','1210599','1210598','1210609','1210607','1210597']
```

```python
 for i in ID:
        url = ("https://www.espncricinfo.com/ci/content/squad/"+i
               +".html")
        page = requests.get(url)
        print(url)
        soup = BeautifulSoup(page.content,'html.parser')
        list(soup.children)
        right_table=soup.find('ul', class_='large-block-grid-2 medium-block-grid-2 small-block-grid-1')

        A=[]
        B=[]

        counter = 0
        counter1 =0
        for row in right_table.findAll("li"):  
            counter += 1 # Same as counter = counter + 1
            A.append(row.text)
            for link in row.find_all('a'):
                    url = link.get('href')
                    B.append(url)

        print('Number of iterations: ' + str(counter))
        print('Number of iterations1: ' + str(counter1))

        import  pandas as pd

        df = pd.DataFrame([sub.split(",") for sub in A])
        print(df)
        df.to_excel("tempipl.xlsx")

        # open the file
        df1 = pd.read_excel(r"E:\dsp imarticus\tempipl.xlsx")
        df1= df1.rename(columns={'Unnamed: 0':'a',0 :'b'})

        import re
        df1 = df1.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\n',  ' ', regex=True)
        df1 = df1.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\t',  ' ', regex=True)

        df1.to_excel("stcsk2.xlsx")

        df2 = pd.read_excel(r"E:\dsp imarticus\stcsk2.xlsx")
        df3 = df2.b
        df3= df2['b'].str.split(':',expand=True)
        df4= df3[0].str.split('\r',expand=True)
        df4['Name'] =df4[0]
        df5 = df3[1].str.split('Playing role',expand=True)
        df5['Age'] =df5[0]
        df6 = df3[2].str.split(' ',expand=True)
        df6['Role'] =df6[1]
        df7 = df3[3].str.split(' ',expand=True)
        df7['Bat'] =df7[1]

        df_team1 = pd.concat([df4,df5,df6,df7,df3[4]],axis=1)
        df_team2= df_team1[['Name','Age','Role','Bat',4]]
        df_team2= df_team2.rename(columns={4 :'Bowl'})

        df8 = pd.DataFrame([sub.split(",") for sub in B])
        df9= df8.drop_duplicates()
        df10 = df9.reset_index()
        df10 = df10.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\n',  ' ', regex=True)
        df10 = df10.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\t',  ' ', regex=True)
        df10['ResultingColumn'] = ['https://www.espncricinfo.com'+str(x) for x in df10[0]]
        df_team2['playerlink'] = df10[['ResultingColumn']]
        df_team2.head()
        #generate a unique file name based on the id and record
        file_name="id_"+str(i)+".xlsx"
        df_team2.to_excel(file_name,index=False)
```
